###  Day_train변수들을 그대로 사용하여 MonthlyIncome 와 NumberOfDependents 변수 예측한다.
1. train 내부에서 결측치를 예측한다.
2. train 전체 데이터로 학습한 모델으로 test의 결측치를 예측한다.

```python
# train 내부 예측
''' 
1.NumberOfDependents 의 결측률이 2.6%로 상당히 낮다
2.그러므로 NumberOfDependents 결측치를 먼저 제거하여  MonthlyIncome 를 예측한다.
3.재결합 시키기 위해 train의 Unnamed:0 를 사용한다.
'''
train=X
T_train=train['SeriousDlqin2yrs']
F_train=train.drop('SeriousDlqin2yrs',axis=1)
```

### train(F_train)내에서 MonthlyIncome 예측

1.처음에는 MonthlyIncome를 제외한 변수들로 NumberOfDependents를 예측하여 사용할까 했음

2.그러나 모델에는 최대한 부정확한 값들이 없는게 좋겠다고 판단

3.위에서 NumberOfDependents의 결측치가 없어도 변수들간의 상관관계는 크게 변화가 없음을 확인

4.NumberOfDependents 결측값을 포함한 행들은 제거하여 train내에서 MonthlyIncome을 예측

```python
# train에서 NumberOfDependents 결측치를 제거한 새로운 데이터 Mon생성
Mon=train[(train['NumberOfDependents'].isnull()==0)]
```

```python
#Mon데이터를 train과 test로 분리
Mon_train=Mon[(Mon['MonthlyIncome'].isnull()==0)]
Mon_test=Mon[(Mon['MonthlyIncome'].isnull()==1)]

print('Mon_train과 Mon_test의 형태:',Mon_train.shape,Mon_test.shape)
Mon.info()
```

```python
#회귀를 사용해야하기에 데이터를 Mon_train에 맞춰 scaling해준다.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(Mon_train)

Mon_train_scaled=scaler.transform(Mon_train)
Mon_train_scaled=pd.DataFrame(data=Mon_train_scaled,columns=Mon_train.columns)

Mon_test_scaled=scaler.transform(Mon_test)
Mon_test_scaled=pd.DataFrame(data=Mon_test_scaled,columns=Mon_test.columns)
```

```python
print(Mon_train_scaled['SeriousDlqin2yrs'].unique(),Mon_test_scaled['SeriousDlqin2yrs'].unique())
```

### scaling을 진행한 후 Target값은 1->3.65...  0->-0.27...  로 전환되었다.

```python
#모델 적용을 위한 Target과 Feature분리
y_Mon_train=Mon_train_scaled['MonthlyIncome']

X_Mon_train=Mon_train_scaled.drop('MonthlyIncome',axis=1)
X_Mon_train1=X_Mon_train.drop('Unnamed: 0',axis=1)  #ID를 제거해준다
X_Mon_train1=X_Mon_train1.drop('SeriousDlqin2yrs',axis=1)    #Target값이 변수에 영향을 주지 않도록 제거한다

X_Mon_test=Mon_test_scaled.drop('MonthlyIncome',axis=1)   
X_Mon_test1=X_Mon_test.drop('Unnamed: 0',axis=1)   #ID를 제거해준다 
X_Mon_test1=X_Mon_test1.drop('SeriousDlqin2yrs',axis=1)   #Target값이 변수에 영향을 주지 않도록 제거한다
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


#선형회귀 정확도 체크를 위해서 Mon_train를 새롭게 분리하여 회귀를 돌려본다. 
X_train,X_test,y_train,y_test=train_test_split(X_Mon_train1,y_Mon_train,test_size=0.25,random_state=156)


lr=LinearRegression()
lr.fit(X_train,y_train)
y_preds=lr.predict(X_test)
mse=mean_squared_error(y_test, y_preds)
rmse=np.sqrt(mse)

print("MSE & RMSE :",mse,rmse)
```

### RMSE가 상당히 낮으므로 단순회귀를 적용해 train data내의 MonthlyIncome을 예측하여 사용한다.


```python
#단순회귀로 train data내의 MonthlyIncome을 예측
lr=LinearRegression()
lr.fit(X_Mon_train1,y_Mon_train)
y_preds=lr.predict(X_Mon_test1)
```

### X_Mon_train1과 X_Mon_test1에 각각 y_Mon_train과 y_preds를 추가한 뒤 결합하여 새로운 train데이터를 생성

```python
X_Mon_train1['MonthlyIncome']=y_Mon_train
X_Mon_test1['MonthlyIncome']=y_preds
X_Mon_train1['ID']=X_Mon_train['Unnamed: 0']   #데이터 결합 후 원래 자리를 sort로 찾기위하여 ID 다시 추가
X_Mon_test1['ID']=X_Mon_test['Unnamed: 0']    #데이터 결합 후 원래 자리를 sort로 찾기위하여 ID 다시 추가
X_Mon_train1['Target']=Mon_train_scaled['SeriousDlqin2yrs']   #최종 target값 추가
X_Mon_test1['Target']=Mon_test_scaled['SeriousDlqin2yrs']     #최종 target값 추가

New_train=X_Mon_train1.append(X_Mon_test1)




#원래 순서대로
New_train.sort_values(by=['ID'],axis=0)
```

원래 ID를 데이터 프레임에 추가하니 결측치가 생겨 scaling되어있는 ID사용
